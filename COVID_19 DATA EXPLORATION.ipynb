# Covid 19 cases by location

```python
covid_cases_by_location=pd.read_sql_query('''
                select location, date, total_cases, new_cases, total_deaths, population
                from Covid_Deaths
                Where continent is not null
                order by location
                ''',conn)

covid_cases_by_location
```

# Total Cases vs Total Deaths

```python
# Shows likelihood of dying if you contract covid in your country
TotalCases_vs_TotalDeaths= pd.read_sql_query('''
                                                Select Location, date, total_cases,total_deaths, 
                                                (total_deaths/total_cases)*100 as DeathPercentage
                                                From Covid_Deaths
                                                Where location like '%canada%' and continent is not null 
                                                order by location
                                                ''', conn)
TotalCases_vs_TotalDeaths
```

# Total Cases vs Population

```python
# Shows what percentage of population infected with Covid
TotalCases_vs_Population= pd.read_sql_query('''
                                               Select Location, date, Population, total_cases, 
                                               (total_cases/population)*100 as PercentPopulationInfected
                                                From Covid_Deaths
                                                Where location like '%canada%'
                                                order by 1
                                                ''', conn)
TotalCases_vs_Population
```

# Countries with Highest Infection Rate compared to Population

```python
# Shows Countries with Highest Infection Rate
Countries_with_HighestInfectionRate= pd.read_sql_query('''
                                                       Select Location, Population, 
                                                       MAX(total_cases) as HighestInfectionCount, 
                                                       Max((total_cases/population))*100 as PercentPopulationInfected
                                                       From Covid_Deaths                                                       
                                                       Group by Location, Population
                                                       order by PercentPopulationInfected desc
                                                        ''', conn)
Countries_with_HighestInfectionRate
```

# Countries with Highest Death Count per Population

```python
# Shows Countries with Highest Death Count per Population
Countries_with_HighestDeathCount=pd.read_sql_query('''
                                                    Select Location, 
                                                    MAX(cast(Total_deaths as int)) as TotalDeathCount
                                                    From Covid_Deaths                                                    
                                                    Where continent is not null 
                                                    Group by Location
                                                    order by TotalDeathCount desc
                                                    ''', conn)
Countries_with_HighestDeathCount
```

# Continent with Highest Death Count per Population

# Total Population vs Vaccinations

```python
# Shows Total Population vs Vaccinations
TotalPopulation_vs_Vaccinations=pd.read_sql_query('''
                                                    Select dea.continent, dea.location, dea.date,
                                                    dea.population, vac.new_vaccinations
                                                    From Covid_Deaths dea
                                                    Join Covid_Vaccinations vac
                                                	On dea.location = vac.location
                                                	and dea.date = vac.date
                                                    where dea.continent is not null 
                                                    order by 2,3
                                                    ''', conn)
TotalPopulation_vs_Vaccinations
```

```python
# Shows Total Population vs Total Vaccinations in each locatio
TotalPopulation_Vs_TotalVacinations=pd.read_sql_query('''
                                                    Select dea.continent, dea.location, 
                                                    dea.date, dea.population, vac.new_vaccinations,
                                                    SUM(vac.new_vaccinations) OVER                                                     
                                                    (Partition by dea.Location Order by dea.location, dea.Date) 
                                                    as RollingPeopleVaccinated
                                                    From Covid_Deaths dea
                                                    Join Covid_Vaccinations vac
                                                	On dea.location = vac.location
                                                    and dea.date = vac.date
                                                    where dea.continent is not null 
                                                    order by 2,3
                                                    ''', conn)
TotalPopulation_Vs_TotalVacinations
```

```python
# Using CTE to perform Calculation on Partition By in previous query
PercentageofPopulation_with_oneCovidVaccine=pd.read_sql_query('''
                                                             With PopvsVac (Continent, Location, Date, Population,
                                                             New_Vaccinations, RollingPeopleVaccinated)
                                                             as
                                                            (
                                                             Select dea.continent, dea.location, dea.date, 
                                                             dea.population, vac.new_vaccinations
                                                             , SUM(vac.new_vaccinations) OVER 
                                                             (Partition by dea.Location Order by dea.location,
                                                             dea.Date) as RollingPeopleVaccinated
                                                             From Covid_Deaths dea
                                                             Join Covid_Vaccinations vac
                                                             On dea.location = vac.location
	                                                         and dea.date = vac.date
                                                             where dea.continent is not null 
                                                             )
                                                            Select *, (RollingPeopleVaccinated/Population)*100 as
                                                            PercentPopulationVaccinated
                                                            From PopvsVac
                                                            ''', conn)
PercentageofPopulation_with_oneCovidVaccine
```

#  Create a Temp Table to get the Percentage Population Vaccinated
# Using temp table to perform Calculation on Partition

```python
conn.execute('''
            DROP Table if exists PercentPopulationVaccinated
            ''')
```

```python
conn.execute('''
             
             Create Table PercentPopulationVaccinated
             (
             Continent nvarchar(255),
             Location nvarchar(255),
             Date datetime,
             Population float,
             New_vaccinations float,
             RollingPeopleVaccinated float
             )
             ''')
```

```python
conn.execute('''
             Insert into PercentPopulationVaccinated
             Select dea.continent, dea.location, dea.date, dea.population, vac.new_vaccinations, 
             SUM(vac.new_vaccinations) OVER (Partition by dea.Location Order by dea.location, dea.Date) 
             as RollingPeopleVaccinated
             From Covid_Deaths dea
             Join Covid_Vaccinations vac
             On dea.location = vac.location
             and dea.date = vac.date
             where dea.continent is not null 
             ''')                                          

                                                
```

```python
PercentPopulationVaccinated=pd.read_sql_query('''
                                              Select *, (RollingPeopleVaccinated/population)*100 
                                              as  PercentPopulationVaccinated
                                              From PercentPopulationVaccinated                                                
                                              ''', conn)
PercentPopulationVaccinated
```

```python
conn.execute('''
            Create View PercentagePopulationVaccinated as
            Select dea.continent, dea.location, dea.date, dea.population, vac.new_vaccinations
            , SUM(vac.new_vaccinations) OVER (Partition by dea.Location Order by dea.location, dea.Date) as RollingPeopleVaccinated
            From Deaths dea
            Join Covid_Vaccinations vac
        	On dea.location = vac.location
        	and dea.date = vac.date
            where dea.continent is not null 
            ''')
```
