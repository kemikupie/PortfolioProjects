# Import software libraries

```python
import pandas as pd
```

# Load the dataset as a `DataFrame`

```python
NY_Citi_Bikes_data=pd.read_csv('Datasets/New York Citi Bikes_Raw Data - NYCitiBikes (1).csv')
```

```python
NY_Citi_Bikes_data
```

```python
NY_Citi_Bikes_data.head(3)
```

# Identify all duplicated data

```python
duplicated_data=NY_Citi_Bikes_data[NY_Citi_Bikes_data.duplicated(keep=False)]

print('number of rows with duplicated data:', 
      duplicated_data.shape[0])
```

```python
duplicated_data
```

# Remove the duplicated data

```python
NY_Citi_Bikes_data_wt_duplicates=NY_Citi_Bikes_data[~NY_Citi_Bikes_data.duplicated()]
```
